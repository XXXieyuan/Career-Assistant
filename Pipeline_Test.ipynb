```python
from selenium import webdriver
from selenium.webdriver.chrome.options import Options
from selenium.webdriver.common.by import By
from selenium.webdriver.support.ui import WebDriverWait
from selenium.webdriver.support import expected_conditions as EC

# Configure Chrome to run headless (no browser window)
chrome_options = Options()
chrome_options.add_argument("--headless")

# Initialize the WebDriver (change the path to your ChromeDriver if needed)
driver = webdriver.Chrome(options=chrome_options)

# URL of the job search page
url = "https://jobs.careers.microsoft.com/global/en/search?l=en_us&pg=1&pgSz=20&o=Relevance&flt=true&ref=cms&utm_medium=Google(MX_FY25DCT)"

driver.get(url)

# Wait until job listings appear. We look for an element that only appears when jobs are loaded.
# In this case, we're using a CSS selector that matches each job post cell.
wait = WebDriverWait(driver, 20)  # Timeout after 20 seconds if not found
job_elements = wait.until(
    EC.presence_of_all_elements_located((By.CSS_SELECTOR, '[data-automationid="ListCell"]'))
)

# Now that we have job_elements, we can extract information.
# Each job cell likely contains an H2 or some heading for the job title.
job_titles = []
for job_el in job_elements:
    # Adjust the selector to target the job title element accurately.
    # For example, if the job title is in an h2 tag, you might do:
    title_el = job_el.find_element(By.TAG_NAME, "h2")
    job_titles.append(title_el.text.strip())

# Print out the extracted job titles
for idx, title in enumerate(job_titles, start=1):
    print(f"{idx}. {title}")

driver.quit()

```

```python
from selenium import webdriver
from selenium.webdriver.chrome.options import Options
from selenium.webdriver.common.by import By
from bs4 import BeautifulSoup
import re

chrome_options = Options()
chrome_options.add_argument("--headless")
driver = webdriver.Chrome(options=chrome_options)

url = "https://jobs.careers.microsoft.com/global/en/search?l=en_us&pg=1&pgSz=20&o=Relevance&flt=true&ref=cms"
driver.get(url)

# Wait for page to load job cells
driver.implicitly_wait(10)  # Simple implicit wait; consider WebDriverWait for more reliability

# Locate all job cells
job_cells = driver.find_elements(By.CSS_SELECTOR, '[data-automationid="ListCell"]')

search_list=[]
for job_cell in job_cells:
    # Get the HTML of the cell to parse with BeautifulSoup
    cell_html = job_cell.get_attribute('outerHTML')
    soup = BeautifulSoup(cell_html, 'html.parser')

    # Extract the job number from the aria-label attribute.
    # For example: aria-label="Job item 1791749"
    job_div = soup.select_one('div[aria-label^="Job item"]')
    
    if not job_div:
        continue  # Skip if no job div found (just a safety check)

    aria_label = job_div.get('aria-label', '')
    # Assuming the format is always "Job item 1791749", split by space and take the last part
    parts = aria_label.split()
   
    job_number = parts[-1] if parts else None
   
    # Extract the job name from the h2 tag with class "MZGzlrn8gfgSs8TZHhv2"
    title_el = soup.select_one('h2.MZGzlrn8gfgSs8TZHhv2')
   
    if not title_el:
        continue  # If for some reason no title is found, skip

    job_name = title_el.get_text(strip=True)
    print(job_name)
    # The URL pattern requires replacing spaces with dashes
    # Example: "Senior Software Engineer" -> "Senior-Software-Engineer"
    job_name_url = re.sub(r'\s+', '-', job_name)
   
    # Construct the URL
    job_url = f"https://jobs.careers.microsoft.com/global/en/job/{job_number}/{job_name_url}"
    search_list.append(job_url)
    print(f"Job Number: {job_number}")
    print(f"Job Name: {job_name}")
    print(f"URL: {job_url}")
    print("-" * 50)

driver.quit()
```

```python
from selenium import webdriver
from selenium.webdriver.chrome.options import Options
from selenium.webdriver.common.by import By
from selenium.webdriver.support.ui import WebDriverWait
from selenium.webdriver.support import expected_conditions as EC
from bs4 import BeautifulSoup
import re
import time

# STEP 1: Extract job URLs from the search page

chrome_options = Options()
chrome_options.add_argument("--headless")
driver = webdriver.Chrome(options=chrome_options)

url = "https://jobs.careers.microsoft.com/global/en/search?lc=Australia&l=en_us&pg=1&pgSz=20&o=Recent"
driver.get(url)

wait = WebDriverWait(driver, 20)
wait.until(EC.presence_of_all_elements_located((By.CSS_SELECTOR, '[data-automationid="ListCell"]')))

job_cells = driver.find_elements(By.CSS_SELECTOR, '[data-automationid="ListCell"]')

search_list = []
for job_cell in job_cells:
    cell_html = job_cell.get_attribute('outerHTML')
    soup = BeautifulSoup(cell_html, 'html.parser')

    # Extract job number from aria-label="Job item XXXXXX"
    job_div = soup.select_one('div[aria-label^="Job item"]')
    if not job_div:
        continue
    aria_label = job_div.get('aria-label', '')
    parts = aria_label.split()
    job_number = parts[-1] if parts else None
    if not job_number:
        continue

    # Extract job name from h2 tag (adjust the class if needed)
    title_el = soup.select_one('h2.MZGzlrn8gfgSs8TZHhv2')
    if not title_el:
        continue
    job_name = title_el.get_text(strip=True)

    # Format job name for URL (replace spaces with dashes)
    job_name_url = re.sub(r'\s+', '-', job_name)
   
    # Construct the job detail URL
    job_url = f"https://jobs.careers.microsoft.com/global/en/job/{job_number}/{job_name_url}"
    search_list.append(job_url)

driver.quit()

# STEP 2: Navigate to each job detail page and extract details

chrome_options = Options()
chrome_options.add_argument("--headless")
driver = webdriver.Chrome(options=chrome_options)
wait = WebDriverWait(driver, 30)

job_details_list = []
error_list=[]
for job_url in search_list:
    time.sleep(10)
    driver.get(job_url)
    try:
        # Wait for a stable element on the job detail page.
        # Inspect the actual page to find a guaranteed element. 
        # Example: Let's assume there's always an H1 with class 'job-title' on detail page:
        detail_element = wait.until(
            EC.presence_of_element_located((By.CSS_SELECTOR, 'div[class*="SearchJobDetailsCard"]'))  # Adjust this selector
        )
        
        # Get the entire page source now that we know the detail page is loaded
        detail_html = driver.page_source
        print(job_url)
        job_details_list.append(detail_html)
    except:
        # If we fail to find the element, print the page source for debugging
        # This helps identify what selector to use in the future.
        print(f"Could not find the detail element for {job_url}.")
        error_list.append(job_url)
        print("-" * 50)
        # Append the raw HTML anyway, so we can later inspect it outside the script
        job_details_list.append(driver.page_source)

driver.quit()

# # STEP 3: Now 'job_details_list' contains the HTML of each job's detail page.
# # You can parse each HTML with BeautifulSoup as needed:
# for i, details_html in enumerate(job_details_list, start=1):
#     soup = BeautifulSoup(details_html, 'html.parser')
#     # Extract something from the detail page (adjust selectors based on actual structure)
#     # For example, using the previously found job title element:
#     title_el = soup.select_one('h1.job-title-selector')  # Adjust this selector
#     title = title_el.get_text(strip=True) if title_el else "N/A"

#     print(f"Job {i} Title: {title}")
#     print("-" * 50)
```
